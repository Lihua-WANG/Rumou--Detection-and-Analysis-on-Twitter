```python
def get_labels(file1, file2, file3):
  with open(file1, 'r') as f:
      id, label1 = [], []
      temp1 = json.loads(f.read())
      for key, val in temp1.items():
        id.append(str(key))
        label1.append(val)
      for i in range(len(label1)):
          label1[i] = 1 if label1[i] == "rumour" else 0 #convert_label(label_data["label"][i])
      df1 = pd.DataFrame(label1, columns=['label1'])
      # df_id = pd.DataFrame(id, columns=['id'])
  with open(file2, 'r') as f:
      label2 = []
      temp2 = json.loads(f.read())
      for key, val in temp2.items():
        label2.append(val)
      for i in range(len(label2)):
          label2[i] = 1 if label2[i] == "rumour" else 0 
      df2 = pd.DataFrame(label2, columns=['label2'])
  # df3=pd.DataFrame(l3, columns=['label3'])
  with open(file3, 'r') as f:
      label3 = []
      temp3 = json.loads(f.read())
      for key, val in temp3.items():
        label3.append(val)
      for i in range(len(label3)):
          label3[i] = 1 if label3[i] == "rumour" else 0 
      df3 = pd.DataFrame(label3, columns=['label3'])
  df_label = pd.concat([df1, df2, df3],axis=1)
  return id, df_label
```
